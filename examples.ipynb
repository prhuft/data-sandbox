## Adding background map to plots with contextily
https://geopandas.org/en/stable/gallery/plotting_basemap_background.html

```python
df = geopandas.read_file(geopandas.datasets.get_path('nybb'))
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
```

```python
df.crs
```

```python
# Contextily uses web map tiles which use ESPG=3857
df_wm = df.to_crs(epsg=3857)
ax = df_wm.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax)
```

```python
# instead of adjusting the crs of the data,
# we can just pass the geopandas df's crs to 
# contextily

ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, crs=df.crs)
```

```python
# adjust the detail using zoom
ax = df_wm.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, zoom=12)
```

```python
# differnt style:
ax = df_wm.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, source=cx.providers.Stamen.TonerLite)
ax.set_axis_off()
```
