```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
sx = 1
sy = 2
x = random.normal(0,sx,5000)
y = random.normal(0,sy,5000)

rot = lambda theta: array([[cos(theta), sin(theta)],[-sin(theta), cos(theta)]])

coords = array([x,y]) #.transpose()
coords = dot(rot(0.4),coords)
x,y = coords

fig,ax = plt.subplots()
ax.scatter(x,y,s=0.5)
ax.set_xlim((-8,8))
ax.set_ylim((-8,8))
ax.set_aspect('equal')

# wikipedia gives a 2 x 2 covariance matrix for this data despite the fact that the cross-covariance matrix for two 
# 1d arrays of length n is n x n. How did they get this? 
```
